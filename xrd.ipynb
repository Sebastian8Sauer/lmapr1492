```python
from pymatgen.ext.matproj import MPRester
with MPRester("Os4JHJeA2G7FbSIy4") as m:
        
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-568662")
    
    print("Le matériau analysé est le BaCl2")
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
# On impose la longueur d'onde du rayonnement CuKa : 1,5406 angstroms
# Car sa valeur standard trouve dans le code source de AVAILABLE_RADIATION n'est pas egale a la valeur souhaite
xrd = XRDCalculator(1.5406)
# Obtention du pattern de diffraction pour notre materiau
pattern = xrd.get_pattern(structure)
```

```python
#Affichage des premiers pics du diffractogramme (pour 2*theta allant de 0 à 45°)
xrd.show_plot(structure, two_theta_range=[0,45])
```

```python
# Les donnees du diagramme
two_theta = pattern.x
intensity = pattern.y
hkls = pattern.hkls

print("Voici les 3 premiers pics du diffractogramme correspondant (en indiquant les indices hkl associés)") 
print("pour un rayonnement X incident CuKa d'une longueur d'onde de 1,5406 Å","\n")
print("Note : J'exprime l'intensité en pourcentage par rapport à l'intensité du rayon incident")
```

```python
# Le premier pic de diffraction
print("Les données pour le premier pic de diffraction sont : ", "\n")

print("Angle 2\u0398 = ", two_theta[0],"°")
print("Intensité = ", intensity[0],"%")
print("Indices hkl = ", hkls[0],'\n')
```

```python
# Le second pic de diffraction
print("Les données pour le second pic de diffraction sont : ", "\n")

print("Angle 2\u0398 = ", two_theta[1],"°")
print("Intensité = ", intensity[1],"%")
print("Indice hkl = ", hkls[1], '\n')
```
